```python

!pip install -U lightrag[ollama]
# Placeholder for lightRAG and Ollama imports
from lightrag.components.model_client import OllamaClient
from lightrag.core.generator import Generator
from lightrag.core.component import Component
from lightrag.core.model_client import ModelClient

import time

# Define QA Component
qa_template = r"""<SYS>
You are a legal expert proficient in constitutional law and case law.
</SYS>
User:  Based on the given article "{{article_id}}", perform the following tasks:
1. Identify significant legal cases related to the article from
the indian constitution but don't display the identified cases and you have to
print according to the prompts given of the task 3 else don't print .
2. From the identified cases, randomly select one case.
3. For the selected case:
   - Provide a brief introduction to the case.
   - Describe the case proceedings clearly and engagingly.
   - Generate at least five dynamic and interactive questions about the case.
     - Questions should be analytical, predictive, or opinion-based.
     - Use a conversational tone to keep the user immersed.
     - End each question with the "#" symbol.
4. Conclude with a summary of the case, key points, and the actual verdict delivered by the court.
You:
"""
class SimpleQA(Component):
    def __init__(self, model_client: ModelClient, model_kwargs: dict):
        super(SimpleQA, self).__init__(model_client, model_kwargs)  # Call super().__init__ with necessary arguments
        self.generator = Generator(
            model_client=model_client,
            model_kwargs=model_kwargs,
            template=qa_template,
        )

    def call(self, input: dict) -> str:
        return self.generator.call({"article_id": str(input)})

    async def acall(self, input: dict) -> str:
        return await self.generator.acall({"article_id": str(input)})

from lightrag.components.model_client import OllamaClient
from IPython.display import Markdown, display

# Initialize Ollama client
model = {
    "model_client": OllamaClient(),
    "model_kwargs": {"model": "llama3.1:8b"}
}
qa = SimpleQA(**model)

# User input for the article
user_article = input("Enter the article (e.g., 'Article 14'): ").strip()

# Pass the article ID to the Ollama model
output = qa.call(user_article)

# Display the output
display(Markdown(f"### Generated Output:\n{output.data}"))
```

```python
from flask import Flask, request, jsonify
from flask_cors import CORS
from pyngrok import ngrok

app = Flask(__name__)  # Changed _name to _name_
CORS(app)  # Enable Cross-Origin Resource Sharing
```

```python
@app.route('/generate', methods=['POST'])
def generate_response():
    # Get the user input from the POST request
    data = request.json
    Prompt = data.get('input', '')

    # Call your model and get the output
    output = qa(Prompt)

    # Extract the generated text from the model's output
    generated_text = output.data

    # Return the response as JSON
    return jsonify({'output': generated_text})
```

```python
# Run the Flask server
from threading import Thread

def run_flask():
    app.run(host='0.0.0.0', port=5000)

thread = Thread(target=run_flask)
thread.start()
```
